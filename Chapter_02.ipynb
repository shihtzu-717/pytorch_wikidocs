#### (2) mean

```python
t = torch.FloatTensor([1, 2])
print(t.mean())
```

```python
t = torch.FloatTensor([[1, 2], [3, 4]])
print(t.mean())
```

```python
t = torch.FloatTensor([[1, 2], [3, 4]])
print(t.mean(dim=0)) # 각 열의 평균값을 구한다 dim=0이라는 것은 t의 첫번째 차원을 제거
```

#### (3) sum

```python
t = torch.FloatTensor([[1, 2], [3, 4]])
print(t)
```

```python
print(t.sum()) # 전체 원소의 덧셈
print(t.sum(dim=0)) # 행을 제거 -> 첫 번째 차원이 바뀌는 위치만 더하면 됨 (0, 0) + (0, 1)
print(t.sum(dim=1)) # 열을 제거 -> (0, 0) + (1, 0)
print(t.sum(dim=-1)) # 열을 제거
```

## 4. 파이썬 클래스

## 1) 함수와 클래스의 차이

```python
result = 0
```

```python
def add(num):
    global result
    result += num
    return result
```

```python
print(add(3))
print(add(4))
```

```python
result1 = 0
result2 = 0

def add1(num):
    global result1
    result1 += num
    return result1

def add2(num):
    global result2
    result2 += num
    return result2

print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))
```

```python
class Calculator:
    def __init__(self):
        self.result = 0
    def add(self, num):
        self.result += num
        return self.result
```

```python
cal1=Calculator()
```

```python
cal2=Calculator()
```

```python
print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```
